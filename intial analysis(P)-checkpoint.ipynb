```python
import pandas as pd
```

```python
customer_shopping_behavior = pd.read_csv("customer_shopping_behavior (1).csv")
customer_shopping_behavior
```

```python
customer_shopping_behavior.isnull().sum()
```

```python
customer_shopping_behavior["Review Rating"] = customer_shopping_behavior.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
customer_shopping_behavior.isnull().sum()
```

```python
customer_shopping_behavior.columns = customer_shopping_behavior.columns.str.lower()
customer_shopping_behavior.columns = customer_shopping_behavior.columns.str.replace(" ","_")
customer_shopping_behavior = customer_shopping_behavior.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
customer_shopping_behavior.columns
```

```python
labels =["young adult","adult","middle age","senior"]
customer_shopping_behavior["age_group"]=pd.qcut(customer_shopping_behavior["age"],q=4,labels = labels)
```

```python
customer_shopping_behavior[["age","age_group"]].head(20)
```

```python
# create a column name purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually': 365,
    'Every 3 Months': 90
}
customer_shopping_behavior["purchase_frequency_days"] = customer_shopping_behavior['frequency_of_purchases'].map(frequency_mapping)
```
